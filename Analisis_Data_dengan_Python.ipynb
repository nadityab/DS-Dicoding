## Cleaning Data

```python
print("Jumlah Duplicated data : ",customers_df.duplicated().sum())
```

```python
customers_df[customers_df.gender.isna()]
```

```python
customers_df.gender.value_counts()
```

```python
customers_df.fillna(value="Prefer not to say",inplace=True)
print('Missing value pada gender sudah diisi "Prefer not to say"')
```

```python
 customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.age.replace(customers_df.age.max(),20,inplace=True)
```
